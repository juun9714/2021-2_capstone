```python
import requests
import json
import time
import matplotlib.pyplot as plt
```

```python
url_items = "http://192.168.50.94:8000/Vehicle/Battery"

cnt = 100
delay_list = []
msg_idx = [i for i in range(1,cnt+1)]

idx = 0
while idx < cnt:
    
    time_before = time.time()
    response = requests.get(url_items)
    if 'r1ID' not in response.text:
        continue
    print(response.text)
    time_after = time.time()
    delay = time_after - time_before
    delay_list.append(delay)
    print(str(idx) + ": "+str(delay))
    print("\n")
    idx += 1
    #time.sleep(0.5)
    
```

```python
plt.plot(msg_idx, delay_list, 'bo-')
plt.ylim([0, 0.2])
plt.xticks(msg_idx)
plt.rcParams['figure.figsize'] = [10, 5]
plt.xlabel('Message Index')
plt.ylabel("Delay(s)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame(delay_list, index=msg_idx, columns=['Delay'])
df.to_csv("delay.csv")
```
